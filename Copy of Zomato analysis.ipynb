## 3. ***Data Wrangling***

### Data Wrangling Code

````python
import pandas as pd

# Load the dataset
zomato_data = pd.read_csv('zomato.csv')

# Display the first few rows of the dataset to understand its structure
print(zomato_data.head())

# Check for missing values
print(zomato_data.isnull().sum())

# Drop rows with missing values, if any
zomato_data.dropna(inplace=True)

# Check for duplicate rows
print("Number of duplicate rows:", zomato_data.duplicated().sum())

# Drop duplicate rows, if any
zomato_data.drop_duplicates(inplace=True)

# Convert 'Average Cost for two' column to numeric
zomato_data['Average Cost for two'] = pd.to_numeric(zomato_data['Average Cost for two'], errors='coerce')

# Convert 'Has Table booking', 'Has Online delivery', and 'Is delivering now' columns to boolean
boolean_columns = ['Has Table booking', 'Has Online delivery', 'Is delivering now']
for col in boolean_columns:
    zomato_data[col] = zomato_data[col].map({'Yes': True, 'No': False})

# Filter out rows where 'Price range' is not within the range of 1 to 4
zomato_data = zomato_data[(zomato_data['Price range'] >= 1) & (zomato_data['Price range'] <= 4)]

# Reset index after dropping rows
zomato_data.reset_index(drop=True, inplace=True)

# Save the cleaned dataset to a new CSV file
zomato_data.to_csv('cleaned_zomato_data.csv', index=False)

# Display summary statistics of numeric columns
print(zomato_data.describe())

# Display unique values of categorical columns
print(zomato_data['Rating color'].unique())
print(zomato_data['Rating text'].unique())
````

### What all manipulations have you done and insights you found?




* Checked for missing values and dropped rows with missing data.

* Checked for duplicate rows and dropped duplicates.
* Converted 'Average Cost for two' column to numeric.

* Converted 'Has Table booking', 'Has Online delivery', and 'Is delivering now'
  columns to boolean.
* Filtered out rows where 'Price range' is not within the range of 1 to 4.
  




* Reset index after dropping rows.
